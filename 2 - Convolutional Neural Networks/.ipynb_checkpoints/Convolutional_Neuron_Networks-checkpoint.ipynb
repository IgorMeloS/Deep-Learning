**Convolutional Neuron Networks in Python**


# Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

tf.__version__

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from keras.preprocessing.image import ImageDataGenerator 
```

# Data Preprocessing

## Loading the training and test set

Defining some parameters to load the dataset

```python
batch_size = 32
img_height = 180
img_width = 180
```

### Training set

```python
train_set = keras.preprocessing.image_dataset_from_directory(
  'dataset/training_set',
  seed=0,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

### Test set


```python
test_set = keras.preprocessing.image_dataset_from_directory(
  'dataset/training_set',
  seed=0,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

### Class names

```python
class_names = train_set.class_names
print(class_names)
```

### Visualizing the data

```python
plt.figure(figsize=(10, 10))
for images, labels in train_set.take(1):
      for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")
plt.show()
```

## Configure the dataset for performance## 

```python
AUTOTUNE = tf.data.AUTOTUNE

train_set = train_set.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
test_set = test_set.cache().prefetch(buffer_size=AUTOTUNE)
```

## Standardization of the data

```python
normalization_layer = layers.experimental.preprocessing.Rescaling(1./255)
```

```python
normal_train_set = train_set.map(lambda x, y: (normalization_layer(x), y))
image_batch, labels_batch = next(iter(normal_train_set))
first_image = image_batch[0]
# Notice the pixels values are now in `[0,1]`.
print(np.min(first_image), np.max(first_image))
```

# Building the model

## Creating the model

```python
num_classes = 2

cnn_model = Sequential([
  layers.Conv2D(16, 3, padding='same', activation='relu', input_shape=(img_height, img_width, 3)),
  layers.MaxPooling2D(),
  layers.Conv2D(32, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(64, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Flatten(),
  layers.Dense(128, activation='relu'),
  layers.Dense(units=1, activation='sigmoid')
])
```

## Compiling the model

```python
cnn_model.compile(optimizer='adam',
              loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

## Model summary

```python
cnn_model.summary()
```

## Training the model

```python
epochs=10
history = cnn_model.fit(
  train_set,
  validation_data=test_set,
  epochs=epochs
)
```

### Visualizing some metrics score (accuracy and loss)

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
data_augmentation = keras.Sequential(
  [
    layers.experimental.preprocessing.RandomFlip("horizontal", 
                                                 input_shape=(img_height, 
                                                              img_width,
                                                              3)),
    layers.experimental.preprocessing.RandomRotation(0.1),
    layers.experimental.preprocessing.RandomZoom(0.1),
  ]
)
```

## Dropout


```python
cnn_model_dr = Sequential([
  data_augmentation,
  layers.Conv2D(16, 3, padding='same', activation='relu', input_shape=(img_height, img_width, 3)),
  layers.MaxPooling2D(),
  layers.Conv2D(32, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(64, 3, padding='same', activation='relu'),
  layers.Dropout(0.2),
  layers.MaxPooling2D(),
  layers.Flatten(),
  layers.Dense(128, activation='relu'),
  layers.Dense(units=1, activation='sigmoid')
])
```

```python
cnn_model_dr.compile(optimizer='adam',
              loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
cnn_model_dr.summary()
```

```python
epochs = 15
history_dr = cnn_model_dr.fit(
  train_set,
  validation_data=test_set,
  epochs=epochs
)
```

```python
acc = history_dr.history['accuracy']
val_acc = history_dr.history['val_accuracy']

loss = history_dr.history['loss']
val_loss = history_dr.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

## Making a single prediction

```python
test_image = keras.preprocessing.image.load_img('dataset/single_prediction/scarlatte2.jpeg',
                                               target_size=(img_height, img_width))
test_image_array = keras.preprocessing.image.img_to_array(test_image)
test_image_array = tf.expand_dims(test_image_array, 0)
prediction = cnn_model.predict(test_image_array)

score = tf.nn.softmax(prediction[0])
print(
    "This image most likely belongs to {} with a {:.2f} percent confidence."
    .format(class_names[np.argmax(score)], 100 * np.max(score))
)
```
